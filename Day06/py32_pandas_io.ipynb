##### 결측치 확인, 제거, 채우기

- 데이터 분석 전에 해야 하는 가장 중요한 일

```python
# 결측치 확인
datas = {
    '이름' : ['길동이', '길순이', '유고', '애슐리'],
    '국어' : [90, 100, None, 99],
    '영어' : [None, 100, 99, 100],
    '수학' : [50, 100, None, 20],
}
df5 = pd.DataFrame(datas)
df5
```

```python
df5.isnull()
```

```python
df5.isnull().sum()
```

```python
df5.notnull().sum() # 잘 안씀
```

```python
# 결측치 채우기 => 재할당 필요
df5['국어'] = df5['국어'].fillna(50)
df5
```

```python
# inplace=True => 재할당 필요 X
df5['영어'].fillna(40, inplace=True)
df5
```

```python
# method='ffill' => foreward 값으로 채워넣기
df5['수학'].fillna(method='ffill')
```

```python
# method='bfill' => backword 값으로 채워넣기
df5['수학'].fillna(method='bfill', inplace=True)
df5
```

```python
# 결측치 삭제
df6 = pd.DataFrame(datas)
df6
```

```python
# 결측치 삭제
df6.dropna()
```

```python
# 결측치 삭제
df6.dropna(inplace=True)
df6
```

```python
# 결측치 삭제
df7 = pd.DataFrame(datas)
df7
```

```python
df7.dropna(axis=1) # null 이 있는 컬럼 삭제
```

```python
# 형식변환
df7
```

```python
df7['국어']
```

```python
df7['국어'].astype('object')
```

```python
# 결측치 채운 후 타입변경
df7['국어'].fillna(30).astype('int64')
```
